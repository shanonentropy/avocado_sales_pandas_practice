```python
###### model building #########


'''
what I would like to do is have a model that will predict the region given
avocado cosumption (total volume, type and average_price)  
'''


df_model = df[['region','total_volume', 'type', 'avg_price']]

# map 0,1 to "type"

df_model['type_num'] = df_model.type.map({'conventional':0, 'organic':1})

# define the variables
X = df_model[['total_volume', 'type_num', 'avg_price']]
y = df_model['region']


'''build a KNN model that given total_volume, type nad price of avocados 
predicts the region
'''
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import make_pipeline
from sklearn.metrics import confusion_matrix

# baseline KNN model
'''since each region has equal representation in the dataset, baseline is 16%'''

#instantiate the model

model = KNeighborsClassifier(n_neighbors=5, weights='distance')

#fit the model

model.fit(X,y)

#predict

model.predict([X.iloc[0,:]])


```

```python
y.head(1)
```

```python
#define the pipeline, in this case standardscaler and knn

pipe = make_pipeline(StandardScaler(), KNeighborsClassifier())

# define the search range for each of the hyperparameters to be optimized

n_neighbor_range = np.arange(1,50)
weights_range = ['distance', 'uniform']


# set up the param_grid

param_grid = dict(kneighborsclassifier__n_neighbors = n_neighbor_range, 
                  kneighborsclassifier__weights = weights_range)

# gridsearch

c_v = []
params = []
score =[]

for x in range(2,20):
    grid = GridSearchCV(pipe, param_grid, cv=x, scoring='accuracy')
    # fit 
    grid.fit(X, y)
    c_v.append(x)
    params.append(grid.best_params_)
    score.append(grid.best_score_)
```

```python
scoring = pd.DataFrame(list(zip(c_v, params, score)), columns = ['cv', 'params', 'score']); scoring
```
